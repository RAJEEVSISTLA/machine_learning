```python
import matplotlib.pyplot as plt

# Select a feature
feature = data['embed_0']

# Plot histogram
plt.hist(feature, bins=30, alpha=0.7, color='blue')
data['embed_0'].plot(kind='hist')

plt.title('Density Plot for embed_0')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Calculate mean and variance
mean_feature = np.mean(feature)
variance_feature = np.var(feature)
print(mean_feature)
print(variance_feature)
```

```python
from scipy.spatial import distance

# Select two feature vectors
vec1 = data.iloc[0, :-1].values
vec2 = data.iloc[1, :-1].values

# Calculate Minkowski distance for r = 1 to 10
minkowski_distances = [distance.minkowski(vec1, vec2, r) for r in range(1, 11)]

# Plot the Minkowski distance
plt.plot(range(1, 11), minkowski_distances, marker='o')
plt.title('Minkowski Distance between Two Vectors')
plt.xlabel('r')
plt.ylabel('Distance')
plt.show()
```
